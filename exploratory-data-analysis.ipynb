```python
def map_zipcode_to_coords(zipcode):
    try:
        zip_results = zipcodes.matching(str(int(zipcode)).zfill(5))
        return float(zip_results[0]['lat']), float(zip_results[0]['long'])
    except:
        print(f"No Match: {zipcode}")
        return -1, -1


df['lat'], df['lng'] = zip(*df['zip'].progress_apply(map_zipcode_to_coords))
```

```python
variable_one = 'inpatient_beds_used_7_day_avg'
variable_two = 'inpatient_beds_7_day_avg'

df_map = df[df['lat'] > 0]
# df_map = df_map[df_map['collection_week'] == '2020-12-04']
df_map = df_map[df_map['inpatient_beds_used_7_day_avg'] >= 0]
df_map = df_map[df_map['inpatient_beds_7_day_avg'] >= 0]
df_map['proportion_used'] = ((df_map['inpatient_beds_used_7_day_avg'] / df_map['inpatient_beds_7_day_avg']) * 100).astype(int)
df_map['zip'] = df_map['zip'].apply(lambda zipcode: str(int(zipcode)).zfill(5))
df_map = df_map.sort_values(by=['collection_week'])

len(df_map)
```
